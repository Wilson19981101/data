```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen

data1 = dataset.iloc[: , 0:12]

coint_johansen(data1,-1,1).eig

def is_pos_def(coint_johansen(data1,-1,1).eig):
    return np.all(np.linalg.eigvals(coint_johansen(data1,-1,1).eig) > 0)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, dpi=120, figsize=(10,6))
for i, p in enumerate(axes.flatten()):
    data = dataset[dataset.columns[i]]
    p.plot(data, color='blue', linewidth=1)
    
    p.set_title(dataset.columns[i])
    p.xaxis.set_ticks_position('none')
    p.yaxis.set_ticks_position('none')
    
    p.spines["top"].set_alpha(0)
    p.tick_params(labelsize=6)
    
plt.tight_layout();
```
